Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

```python
import re
input_sentence=input("Enter the string: ")
#quoted_input = f"{txt}"
# Split the sentence while preserving repeated words
split_sentence = re.findall(r'\b\w+\b', input_sentence)
c={}
for i in split_sentence:
    if i not in c:
          c[i]=1
    else:
          c[i]=c[i]+1
# Find the highest frequency
max_freq = max(c.values())
print("The words with frequency are : " , c)
print("Highest-frequency of word:", max_freq)

#length of the highest-frequency word
max_freq_word=[word for word, freq in c.items() if freq==max_freq]
print("The max frequent words are : ",max_freq_word)
max_freq_word_length = len(max(max_freq_word, key=len))
print("The length of the word with highest frequency : ", max_freq_word_length)
```

Question 4 -
Write a program to download the data from the link given below and then read the data and convert the into
the proper structure and return it as a CSV file.
Link - https://data.nasa.gov/resource/y77d-th95.json
Note - Write code comments wherever needed for code understanding.

```python
import requests
import csv

def download_data(url):
    response = requests.get(url)  # Send a GET request to the URL
    response.raise_for_status()  # Check if the request was successful

    data = response.json()  # Get the JSON data from the response
    return data

def convert_to_csv(data, filename):
    # Extract the keys from the first dictionary in the list
    keys = data[0].keys()

    with open(filename, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=keys)

        writer.writeheader()  # Write the header row

        for row in data:
            writer.writerow(row)  # Write each row

    print(f"Data saved as {filename}")

# Main program
url = 'https://data.nasa.gov/resource/y77d-th95.json'
filename = 'nasa_data.csv'

# Download the data
data = download_data(url)

# Convert and save as CSV
convert_to_csv(data, filename)
```

Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
def count_words(sentence):
    # Convert the sentence to lowercase
    sentence = sentence.lower()

    # Define the lists of verbs, nouns, pronouns, and adjectives
    verbs = ['am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'do', 'does', 'did', 'can', 'could', 'shall', 'should', 'will', 'would', 'may', 'might', 'must']
    nouns = ['book', 'cat', 'dog', 'house', 'car', 'tree', 'person', 'country', 'city', 'food']
    pronouns = ['i', 'you', 'he', 'she', 'it', 'we', 'they', 'me', 'him', 'her', 'us', 'them']
    adjectives = ['good', 'bad', 'happy', 'sad', 'big', 'small', 'beautiful', 'ugly', 'old', 'new']

    # Initialize counters
    count_verbs = 0
    count_nouns = 0
    count_pronouns= 0
    count_adjectives = 0

    # Split the text into words
    words = sentence.split()

    # Iterate over each word and check whether it is verbs, nouns, pronouns and adjectives
    for word in words:
      if word in verbs:
        count_verbs=count_verbs+1
      elif word in nouns:
            count_nouns=count_nouns+1
      elif word in pronouns:
            count_pronouns =count_pronouns+1
      elif word in adjectives:
            count_adjectives=count_adjectives+1

    # Create a dictionary with the counts
    counts = {
        'verbs': count_verbs,
        'nouns': count_nouns,
        'pronouns': count_pronouns,
        'adjectives': count_adjectives
    }

    return counts







```

```python
sentence=input("Enter the sentence: ")
count=count_words(sentence)
print('The count in a given particular phrase or paragraph:  ' ,count)
```
